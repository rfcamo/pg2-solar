```python
# Dependencies
import pandas as pd
import numpy as np
```

```python
# Import our data into pandas from CSV

data_2021_path = 'data/SGU/2021.csv'
data_2020_path = 'data/SGU/2020.csv'
data_2019_path = 'data/SGU/2019.csv'
data_2018_path = 'data/SGU/2018.csv'
data_2017_path = 'data/SGU/2017.csv'
data_2016_path = 'data/SGU/2016.csv'
data_2015_path = 'data/SGU/2015.csv'
data_2014_path = 'data/SGU/2014.csv'
data_2013_path = 'data/SGU/2013.csv'
data_2012_path = 'data/SGU/2012.csv'
data_2011_path = 'data/SGU/2011.csv'
data_2010_path = 'data/SGU/2010.csv'
data_2009_path = 'data/SGU/2009.csv'
data_2008_path = 'data/SGU/2008.csv'
data_2007_path = 'data/SGU/2007.csv'
data_2006_path = 'data/SGU/2006.csv'
data_2005_path = 'data/SGU/2005.csv'
data_2004_path = 'data/SGU/2004.csv'
data_2003_path = 'data/SGU/2003.csv'
data_2002_path = 'data/SGU/2002.csv'
data_2001_path = 'data/SGU/2001.csv'

sgu_2021_df = pd.read_csv(data_2021_path, low_memory=False)
sgu_2020_df = pd.read_csv(data_2020_path, low_memory=False)
sgu_2019_df = pd.read_csv(data_2019_path, low_memory=False)
sgu_2018_df = pd.read_csv(data_2018_path, low_memory=False)
sgu_2017_df = pd.read_csv(data_2017_path, low_memory=False)
sgu_2016_df = pd.read_csv(data_2016_path, low_memory=False)
sgu_2015_df = pd.read_csv(data_2015_path, low_memory=False)
sgu_2014_df = pd.read_csv(data_2014_path, low_memory=False)
sgu_2013_df = pd.read_csv(data_2013_path, low_memory=False)
sgu_2012_df = pd.read_csv(data_2012_path, low_memory=False)
sgu_2011_df = pd.read_csv(data_2011_path, low_memory=False)
sgu_2010_df = pd.read_csv(data_2010_path, low_memory=False)
sgu_2009_df = pd.read_csv(data_2009_path, low_memory=False)
sgu_2008_df = pd.read_csv(data_2008_path, low_memory=False)
sgu_2007_df = pd.read_csv(data_2007_path, low_memory=False)
sgu_2006_df = pd.read_csv(data_2006_path, low_memory=False)
sgu_2005_df = pd.read_csv(data_2005_path, low_memory=False)
sgu_2004_df = pd.read_csv(data_2004_path, low_memory=False)
sgu_2003_df = pd.read_csv(data_2003_path, low_memory=False)
sgu_2002_df = pd.read_csv(data_2002_path, low_memory=False)
sgu_2001_df = pd.read_csv(data_2001_path, low_memory=False)
```

```python
sgu_2021_df
```

```python
SGU_1_df = pd.merge(sgu_2001_df,sgu_2002_df, on = ["Postcode"],suffixes=('_2001', '_2002'), how = "outer")
SGU_2_df = pd.merge(sgu_2003_df,sgu_2004_df, on = ["Postcode"],suffixes=('_2003', '_2004'), how = "outer")
SGU_3_df = pd.merge(sgu_2005_df,sgu_2006_df, on = ["Postcode"],suffixes=('_2005', '_2006'), how = "outer")
SGU_4_df = pd.merge(sgu_2007_df,sgu_2008_df, on = ["Postcode"],suffixes=('_2007', '_2008'), how = "outer")
SGU_5_df = pd.merge(sgu_2009_df,sgu_2010_df, on = ["Postcode"],suffixes=('_2009', '_2010'), how = "outer")
SGU_6_df = pd.merge(sgu_2011_df,sgu_2012_df, on = ["Postcode"],suffixes=('_2011', '_2012'), how = "outer")
SGU_7_df = pd.merge(sgu_2013_df,sgu_2014_df, on = ["Postcode"],suffixes=('_2013', '_2014'), how = "outer")
SGU_8_df = pd.merge(sgu_2015_df,sgu_2016_df, on = ["Postcode"],suffixes=('_2015', '_2016'), how = "outer")
SGU_9_df = pd.merge(sgu_2017_df,sgu_2018_df, on = ["Postcode"],suffixes=('_2017', '_2018'), how = "outer")
SGU_10_df = pd.merge(sgu_2019_df,sgu_2020_df, on = ["Postcode"],suffixes=('_2019', '_2020'), how = "outer")
```

```python

SGU_12_df = pd.merge(SGU_1_df,SGU_2_df, on = ["Postcode"], how = "outer")
SGU_34_df = pd.merge(SGU_3_df,SGU_4_df, on = ["Postcode"], how = "outer")
SGU_56_df = pd.merge(SGU_5_df,SGU_6_df, on = ["Postcode"], how = "outer")
SGU_78_df = pd.merge(SGU_7_df,SGU_8_df, on = ["Postcode"], how = "outer")
SGU_910_df = pd.merge(SGU_9_df,SGU_10_df, on = ["Postcode"], how = "outer")

SGU_1234_df = pd.merge(SGU_12_df,SGU_34_df, on = ["Postcode"], how = "outer")
SGU_5678_df = pd.merge(SGU_56_df,SGU_78_df, on = ["Postcode"], how = "outer")
SGU_91021_df = pd.merge(SGU_910_df,sgu_2021_df, on = ["Postcode"], how = "outer")

SGU_12345678_df = pd.merge(SGU_1234_df,SGU_5678_df, on = ["Postcode"], how = "outer")

combined_df = pd.merge(SGU_12345678_df,SGU_91021_df, on = ["Postcode"], how = "outer")
combined_df = combined_df.rename (columns ={'Installation' :'Installation_2021',
                             'Output' :'Output_2021'})
combined_df
```

```python
combined_df.to_csv(r'data/SGU/Allyears.csv', index = False, header=True)
```
