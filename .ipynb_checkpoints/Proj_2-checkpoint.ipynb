```python
# Dependencies
import pandas as pd
import numpy as np
```

```python
# Import our data into pandas from CSV

postcode_data_path = 'data/postcode-data.csv'

postcode_df = pd.read_csv(postcode_data_path, low_memory=False)

postcode_df
```

```python
postcode_df= postcode_df.dropna()
postcode_df
```

```python
postcode_df.dtypes
```

```python
postcode_df = postcode_df.groupby('Postcode').agg(Suburb = pd.NamedAgg(column='Suburb',aggfunc ='first'),
                                                 State = pd.NamedAgg(column='State',aggfunc ='first'))
postcode_df
```

```python
# Import our data into pandas from CSV
SGU_data_path = 'data/SGU-SolarPanel.csv'
SGU_df = pd.read_csv(SGU_data_path, low_memory=False)
SGU_df.head()
```

```python
SGU_df = pd.DataFrame(SGU_df.values[2:])
SGU_df
```

```python
# Set the column labels to equal the values in the fisrt row
SGU_df.columns = SGU_df.iloc[0]
SGU_df
```

```python
#  drop the first row
SGU_df = SGU_df.drop(SGU_df.index[0])
SGU_df
```

```python
SGU_df = SGU_df.rename (columns ={'Small Unit Installation Postcode' :'Postcode'})
SGU_df
```

```python
SGU_df.dtypes
```

```python
SGU_df['Postcode'] = SGU_df['Postcode'].apply(pd.to_numeric)
SGU_df.dtypes
```

```python
df = pd.merge(postcode_df,SGU_df,  on = ['Postcode'], how = 'right')
df
```

```python
NAN_df = df[df.isna().any(axis=1)]
NAN_df
```

```python
final_df = df.dropna()
final_df
```
